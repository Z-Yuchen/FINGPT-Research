```python
import sys
sys.path.append('/content/FinNLP/')
from transformers import AutoModel, AutoTokenizer, AutoModelForCausalLM, LlamaForCausalLM, LlamaTokenizerFast
from peft import PeftModel
import torch
from finnlp.benchmarks.fpb import test_fpb
from finnlp.benchmarks.fiqa import test_fiqa , add_instructions
from finnlp.benchmarks.tfns import test_tfns
from finnlp.benchmarks.nwgi import test_nwgi
```

```python
def print_model_result(model,tokenizer):
  res_fpb = test_fpb(model, tokenizer, batch_size = batch_size)
  res_tfns = test_tfns(model, tokenizer, batch_size = batch_size)
  res_fiqa = test_fiqa(model, tokenizer, prompt_fun = add_instructions, batch_size = batch_size)
  res_nwgi = test_nwgi(model, tokenizer, batch_size = batch_size)
  print(res_fpb)
  print(res_tfns)
  print(res_fiqa)
  print(res_nwgi)
```

#Base Model: chatglm2-6b

```python
batch_size = 8
base_model1 = "THUDM/chatglm2-6b"
peft_model = "oliverwang15/FinGPT_v31_ChatGLM2_Sentiment_Instruction_LoRA_FT"
tokenizer1 = AutoTokenizer.from_pretrained(base_model1, trust_remote_code=True)
model1 = AutoModel.from_pretrained(base_model1, trust_remote_code=True, load_in_8bit = True, device_map = "auto")
model1 = PeftModel.from_pretrained(model1, peft_model)
model1 = model1.eval()
print_model_result(model1,tokenizer1)
```
